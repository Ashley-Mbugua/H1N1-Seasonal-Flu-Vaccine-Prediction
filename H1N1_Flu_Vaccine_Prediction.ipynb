```python
import pandas as pd
import numpy as np

# Load the training data
train_features = pd.read_csv('/content/training_set_features.csv')
train_labels = pd.read_csv('/content/training_set_labels.csv')

# Load the test data
test_features = pd.read_csv('/content/test_set_features.csv')

# Load the sample submission format
submission_format = pd.read_csv('/content/submission_format.csv')
```

```python
# Check for missing values in training features
missing_values = train_features.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Combine training and test features to apply the same cleaning
combined_features = pd.concat([train_features, test_features], axis=0)

# Step 1: Fill numerical columns with mean
num_cols = combined_features.select_dtypes(include=['float64', 'int64']).columns
combined_features[num_cols] = combined_features[num_cols].fillna(combined_features[num_cols].mean())

# Step 2: Fill categorical columns with mode
cat_cols = combined_features.select_dtypes(include=['object']).columns
combined_features[cat_cols] = combined_features[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))

# Step 3: Split the cleaned combined dataset back into train and test sets
train_features_cleaned = combined_features.iloc[:len(train_features)]
test_features_cleaned = combined_features.iloc[len(train_features):]
```

```python
# One-hot encode the cleaned features
train_features_encoded = pd.get_dummies(train_features_cleaned)
test_features_encoded = pd.get_dummies(test_features_cleaned)

# Align the columns of train and test so they match perfectly
train_features_encoded, test_features_encoded = train_features_encoded.align(test_features_encoded, join='left', axis=1, fill_value=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Define the labels
labels = train_labels[['h1n1_vaccine', 'seasonal_vaccine']]

# Model for H1N1 Vaccine
rf_h1n1 = RandomForestClassifier(random_state=42)
rf_h1n1.fit(train_features_encoded, labels['h1n1_vaccine'])

# Model for Seasonal Vaccine
rf_seasonal = RandomForestClassifier(random_state=42)
rf_seasonal.fit(train_features_encoded, labels['seasonal_vaccine'])
```

```python
# Make predictions on the test data
h1n1_preds = rf_h1n1.predict_proba(test_features_encoded)[:, 1]
seasonal_preds = rf_seasonal.predict_proba(test_features_encoded)[:, 1]

# Create a DataFrame in the correct submission format
submission = pd.DataFrame({
    'respondent_id': test_features['respondent_id'],
    'h1n1_vaccine': h1n1_preds,
    'seasonal_vaccine': seasonal_preds
})

# Save to CSV
submission.to_csv('my_submission.csv', index=False)
```
